#**Dataset Information**

**Source:** DATA.GOV / https://drive.google.com/drive/folders/1hvQ6qvHQdh22V4EgiaqOqHNkkhf9xNix?usp=sharing

**Year / Timeline:**  Model years ranging from 1999 to 2026

**Dataset Column/features Description:**

* **VIN (1-10)**: The first 10 characters of the Vehicle Identification Number, used as a unique identifier for each vehicle.

* **County**: The county in Washington State where the electric vehicle is registered.

* **City**: The city where the electric vehicle is registered.

* **State**: The state of registration (Washington – WA).

* **Postal Code**: The postal (ZIP) code associated with the vehicle’s registration location.

* **Model Year**: The manufacturing year of the electric vehicle model.

* **Make**: The manufacturer or brand of the electric vehicle (e.g., Tesla, Nissan, Chevrolet).

* **Model**: The specific model name of the electric vehicle.

* **Electric Vehicle Type**: Indicates whether the vehicle is a Battery Electric Vehicle (BEV) or a Plug-in Hybrid Electric Vehicle (PHEV).

* **Clean Alternative Fuel Vehicle (CAFV) Eligibility**: Indicates whether the vehicle qualifies for clean alternative fuel incentives under CAFV regulations.

* **Electric Range**: The maximum distance (in miles) the vehicle can travel on electric power alone.

* **Legislative District**: Legislative district number of the vehicle’s registration area.

* **DOL Vehicle ID**: Unique identifier assigned by the Washington State Department of Licensing.

* **Vehicle Location**: Geographic coordinates (latitude and longitude) of the vehicle’s location.

* **Electric Utility**: Electric utility provider serving the vehicle’s registration area.

## **4. Importing Libraries and Loading the Dataset**

```python
# Importing Librairies

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading the Dataset

url = "https://github.com/ROHAN-PRATHEEP/Data-Analytics-Final-Project/raw/refs/heads/main/Electric_Vehicle_Population_Data.xlsx"
df = pd.read_excel(url)

# Preview of the Dataset
df.head()
```

## **Stage 2 – Data Cleaning and Pre-processing**


## **1. Handling Missing Values**

```python
# Categorical columns: Missing values are replaced with "Unknown".

cat_cols = ['County','City','Vehicle Location','Electric Utility']
df[cat_cols] = df[cat_cols].fillna('Unknown')
```

```python
# Numerical columns: Missing values are filled using the median.

num_cols = ['Postal Code','Electric Range','Legislative District','2020 Census Tract']
df[num_cols] = df[num_cols].fillna(df[num_cols].median())
```

```python
# Missing Value Check

df.isnull().sum()
```

## **2. Handling Duplicate Records**

```python
df.drop_duplicates(inplace = True)
df.duplicated().sum()
```

## **3. Column Renaming and Dropping Unwanted Columns**

```python
# To improve readability and maintain consistency during analysis,
# column names were standardized by converting them to lowercase and replacing spaces with underscores.
# This makes the dataset easier to work with in Python and avoids syntax-related issues.

df.columns = df.columns.str.lower().str.strip().str.replace(' ', '_')
```

```python
df.head()
```

```python
# Dropping Unwanted Columns
# dol_vehicle_id: A unique identifier assigned to each vehicle; not useful for analytical insights.

df.drop(columns=['dol_vehicle_id'], inplace= True)
```

```python
df.head()
```

## **4. Outlier Detection and Treatment**

```python
# Outlier Detection

sns.boxplot(x=df['electric_range'])
```

## **Outlier Treatment:**
Outliers were identified in the electric range variable using boxplots. These extreme values correspond to high-performance electric vehicles and represent valid observations rather than data anomalies. Therefore, the outliers were retained to preserve the true variability of electric vehicle performance in the dataset.

## **5. Skewness Check and Transformation**

```python
# Skewness Check

df['electric_range'].skew()
```

## Skewness = +2.11
* This indicates a strong right (positive) skew

```python
# Apply log transformation

df['electric_range_log'] = np.log1p(df['electric_range'])
df['electric_range_log'].skew()
```

## Skewness = +0.78

* This indicates a moderate right skew

* The distribution is now much closer to symmetric

The original electric range variable exhibited a high positive skewness (2.11). After applying a logarithmic transformation, the skewness was reduced to 0.78, indicating a more balanced distribution suitable for further analysis and visualization.

## **6. Convert data types if needed**
After reviewing the dataset, the original data types were retained as they were consistent with the data source and sufficient for the scope of exploratory data analysis. Identifier-based numerical fields such as postal code, legislative district, and census tract were not used for mathematical operations, and therefore no data type conversion was required.

## **7. Feature Transformations and Derived Fields**

Feature transformation involves creating new variables from existing features to enhance analysis and generate deeper insights. Since the dataset does not contain a full date field, transformations are derived mainly from the **Model Year** and **Electric Range** variables.

```python
# 1. Vehicle Age (Derived Feature)
# To understand how old the electric vehicles are, a new feature called vehicle age is created using the model year.

df['vehicle_age'] = 2026 - df['model_year']
```

```python
# 2. Electric Range Category
# To simplify analysis and enable segmentation, electric range values are grouped into meaningful categories.

df['range_category'] = pd.cut(
    df['electric_range'],
    bins=[0, 100, 200, 300, 1000],
    labels=['Low', 'Medium', 'High', 'Very High']
)
```

## **Stage 3 – EDA and Visualizations**


## A. Univariate Analysis
* Univariate analysis examines the distribution of a single variable.

```python
# 1. Distribution of Electric Vehicle Types

sns.countplot(data=df, x='electric_vehicle_type')
plt.title('Distribution of Electric Vehicle types')
plt.xlabel('Electric Vehicle type')
plt.ylabel('Count')
plt.show()
```

## Interpretation:
Battery Electric Vehicles (BEVs) significantly outnumber Plug-in Hybrid Electric Vehicles (PHEVs), indicating a strong preference for fully electric vehicles in Washington State.

## Business Insight:
This trend suggests growing consumer confidence in charging infrastructure and fully electric technology.

```python
# 2. Distribution of Electric Range

plt.hist(df['electric_range'], bins=10)
plt.title('Distribution of Electric Vehicle Range')
plt.xlabel('Electric Range (miles)')
plt.ylabel('Frequency')
plt.show()

```

## Interpretation:
Most electric vehicles fall within the low to medium range categories, while a smaller number of vehicles exhibit very high electric ranges.

## Business Insight:
The market is dominated by practical-range EVs, while premium high-range vehicles cater to a niche segment.

```python
# 3. Boxplot of Electric Range

sns.boxplot(x=df['electric_range'])
plt.title('Boxplot of Electric Range')
plt.show()
```

## Interpretation:
The boxplot reveals a right-skewed distribution with several high-range outliers.

## Business Insight:
These outliers represent high-performance EV models and highlight technological advancements in battery capacity.

## B. Bivariate Analysis
* Bivariate analysis explores relationships between two variables.

```python
# 1. EV Adoption Trend Over Model Years

year_counts = df['model_year'].value_counts().sort_index()
plt.plot(year_counts.index, year_counts.values)
plt.title('Electric Vechicle Adoption Trend by Model Year')
plt.xlabel('Model Year')
plt.ylabel('Number of Vehicles')
plt.show()
```

## Interpretation:
Electric vehicle adoption shows a strong upward trend in recent years.

## Business Insight:
This growth reflects increased environmental awareness, policy incentives, and improvements in EV affordability.

```python
# 2. Electric Vehicle Type vs Average Electric Range

sns.barplot(
    data=df,
    x='electric_vehicle_type',
    y='electric_range'
)
plt.title('Electric Vehicle Type vs Electric Range')
plt.xlabel('Vehicle Type')
plt.ylabel('Median Electric Range')
plt.show()
```

## Interpretation:
Battery Electric Vehicles have a higher median electric range compared to Plug-in Hybrid vehicles.

## Business Insight:
BEVs are better suited for long-distance travel, making them more attractive for primary vehicle usage.

```python
# 3. Correlation Heatmap (Numerical Variables)

sns.heatmap(
    df[['model_year', 'electric_range', 'vehicle_age']].corr(),
    annot=True,
    cmap='coolwarm'
)
plt.title('Correlation Heatmap')
plt.show()
```

## Interpretation:
Model year shows a positive correlation with electric range, while vehicle age is negatively correlated with range.

## Business Insight:
Newer EV models are technologically superior, offering higher electric ranges.

## C. Multivariate Analysis
* Multivariate analysis examines relationships among three or more variables.

```python
# Electric Range by Vehicle Type and Model Year

sns.lineplot(
    data=df,
    x='model_year',
    y='electric_range',
    hue='electric_vehicle_type',
    estimator='median'
)
plt.title('Electric Range Trend by Vehicle Type')
plt.xlabel('Model Year')
plt.ylabel('Median Electric Range')
plt.show()
```

## Interpretation:
Electric range has consistently increased over time for both BEVs and PHEVs, with BEVs leading.

## Business Insight:
Technological improvements benefit all EV categories, but BEVs gain the most from battery innovation.

```python
# County-wise EV Distribution by Vehicle Type

top_counties = df['county'].value_counts().head(5).index

sns.countplot(
    data=df[df['county'].isin(top_counties)],
    x='county',
    hue='electric_vehicle_type'
)
plt.title('Top Counties by EV Adoption and Type')
plt.xlabel('County')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

## Interpretation:
Urban counties show higher EV adoption, with BEVs dominating across regions.

## Business Insight:
Urban infrastructure and higher charging availability strongly influence EV adoption.

```python
# Pivot Table Analysis – EV Count by County and Type

pd.pivot_table(
    df,
    values='vin_(1-10)',
    index='county',
    columns='electric_vehicle_type',
    aggfunc='count'
).head()
```

## Interpretation:
The pivot table clearly highlights regional differences in EV type preference.

## Business Insight:
Policy and infrastructure planning can be tailored at the county level.
